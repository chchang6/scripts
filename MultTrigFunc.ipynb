```python
x = np.linspace(0, 4*np.pi, 100, endpoint=True)
```

```python
y1 = np.sin(x)
y2 = np.sin(2*x)
#plt.plot(x, y1*y2, x, y1+y2)
plt.plot(x, y2, x, y1*y2)
```

```python
y3 = np.sin(3*x)
plt.plot(x, y1, x, y1*y3)
```

```python
y4 = np.sin(x + np.pi/4)
plt.plot(x, y1, x, y1*y4)
```

```python
y5 = np.sin(2*x + np.pi/4)
plt.plot(x, y1, x, y1*y5)
```
