```python
import numpy as np
import itertools
import math
```

```python
def func1(a, b, c):
    x_range = np.arange(-2.5, 2.6, 0.1)
    y_range = np.arange(-2.5, 2.6, 0.1)
    z = itertools.product(x_range,y_range)
    data = {}
    for i in z:
        x = i[0]
        y = i[1]
        data[(x,y)] = -a*x - b*y - math.sqrt(x**2 + c**2 * y**2)
    t = data.keys()
    x = []; y = []; z = []
    for i in t:
        x.append(i[0])
        y.append(i[1])
        z.append(data[i])
    return np.array(x), np.array(y), np.array(z)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax = fig.gca(projection='3d')
x, y, z = func1(1.,0.,1.)
ax.scatter(x, y, z)
ax.scatter(x, y, np.zeros(len(x)), alpha=0.1)
plt.title('Matplot 3d scatter plot')
plt.show()
```
